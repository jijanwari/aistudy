# ６章　確率

## ６．１　従属と独立

大まかに言うと、Eの発生に関する何らかの情報がFの発生についての情報を与える場合、２つの事象EとFは従属関係にあるといいます  

独立の場合  
P(E,F) = P(E)P(F)  

## ６．２　条件付き確率

両者が独立である必要がないのであれば、「FにおけるEの条件付き確率」は  
P(E|F) = P(E,F)/P(F)  
Fが起きたことを知っている状況でEが起きる確率  
  
P(E,F) = P(F) P(E|F)  
    　　　　↑Fがおきる確率　×　さらにEが起きる確率  

EとFが独立なら P(E|F) = P(E)

２人の子供の例  
一人目　二人目  
女　　　女  
女　　　男  
男　　　女  
男　　　男  

問題１　１人目が女のときに２人とも女の確率は？  
P（２人とも女｜１人目が女）＝　P（２人とも女、１人目が女）／Ｐ（１人目が女）  

ここで、２人共女なら、１人目は必ず女なので、　P（２人共女、１人目が女）＝P（２人共女）  

P（２人とも女｜１人目が女）＝　P（２人とも女）／Ｐ（１人目が女）  ＝ (1/4)/(1/2)

問題２　少なくとも１人が女の場合に、２人とも女である確率は？

P（２人とも女｜１人以上女）＝　P（２人とも女、１人以上女）／Ｐ（１人以上女）  
＝(1/4)/(3/4)  


## ６．３　ベイズの定理

　　　　　0.01の確率で、病気を陰性、病気でないのに陽性と判定する  
  
問題　陽性の人が本当に病気である確率は？  


以下、  
○・・・病気でない  
×・・・病気  
陽・・・陽性
